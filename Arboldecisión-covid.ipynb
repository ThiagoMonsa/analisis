```python
#Importamos las librerias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
# Leer los datos desde un archivo CSV
data = pd.read_csv(r'C:\Users\smons\OneDrive\Escritorio\UNAD\ANALISIS DE DATOS\UNIDAD 3\Cleaned-Data.csv')
```

```python
predictors = ["Contact_Yes","Sore-Throat","Dry-Cough","Contact_No"]
target_col = ['Country']
```

```python
predic = data[predictors]
target = data[target_col]
```

```python
X_train, X_test, y_train, y_test = train_test_split(predic, target, test_size=0.4, random_state=13)
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit(X_train, y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```

```python
# Examinamos la matriz
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(y_test,predicciones)
accuracy
```
